## Aula 01: Fundamentos do HTML5


### O Fluxo de Informações na Internet

* **"Como funciona o fluxo de informações na internet? Pense na sua experiência ao acessar um site."**

        O Endereço (Servidor DNS): O navegador precisa de um endereço IP para saber onde o site está hospedado. O servidor DNS traduz a URL para um IP (ex: `192.168.1.1`).

        O Envio (Servidor Web): O navegador envia a requisição para o servidor web.

        A Resposta (O que é HTML): O servidor web encontra o arquivo `index.html` e o envia de volta ao navegador. **HTML é a linguagem que estrutura o conteúdo dessa resposta.

        A Renderização (Navegador): O navegador recebe o arquivo HTML e "lê" as instruções, transformando o código em uma página visualmente apresentável na tela.

* **Conclusão:** HTML é a espinha dorsal de qualquer site, o esqueleto que define o que é um título, um parágrafo, uma imagem, etc.



### A Estrutura do HTML5

* **Análise:**
    * **`<!DOCTYPE html>`:** Essa linha declara ao navegador qual versão do HTML está sendo utilizada.
    * **`<html lang="pt-br">`:** Explica a tag raiz e o atributo `lang` para definir o idioma do documento.
    * **`<head>`:**
        * É a "cabeça" do documento, onde ficam as informações não visíveis na página.
        * **`<meta charset="UTF-8">`:** Essencial para garantir a exibição correta de caracteres especiais (acentos, cedilha).
        * **`<meta name="viewport" content="width=device-width, initial-scale=1.0">`:** IMPORTANTE para a responsividade em dispositivos móveis.
        * **`<title>`:** Onde o título da página aparece na aba do navegador.
    * **`<body>`:**
        * O "corpo" do documento, onde todo o conteúdo visível da página é inserido.
        * **Estrutura Semântica:**
            * IMPORTANTE para usar tags com significado, não apenas `<div>`.
* **Demonstração de Tags em Ordem:**
    * `<h1>` a `<h6>`: Demonstra a hierarquia dos títulos.
    * `<p>`: Para parágrafos.
    * `<div>`: Use-o para agrupar conteúdo e mostrar a diferença em relação às tags semânticas.
    * `<header>`: Para o cabeçalho do site (logo, navegação).
    * `<nav>`: Para o menu de navegação.
    * `<main>`: O conteúdo principal e único de cada página.
    * `<section>`: Agrupar conteúdo temático.
    * `<article>`: Para conteúdo autônomo (post de blog, notícia).
    * `<footer>`: Para o rodapé (direitos autorais, links sociais).
* **Boas Práticas de Organização e Documentação:**
    * **Organização:**
        * Crie pastas para `css`, `js`, `images`.
        * Sempre use nomes de arquivos em letras minúsculas e sem espaços (ex: `minha-pagina.html`).
    * **Documentação (Comentários):**
        * Demonstre o uso de `` para explicar partes do código.
        * Os comentários são cruciais para a colaboração e para que o próprio desenvolvedor se lembre do que fez no futuro.



### Atividade Prática: Meu Currículo Online

* **Instruções da Atividade:**
    * **Objetivo:** Criar uma página HTML simples com informações pessoais e profissionais.
        * Utilize pelo menos uma tag de título (`h1` ou `h2`).
        * Utilize pelo menos duas tags de parágrafo (`p`).
        * Utilize tags semânticas (`<header>`, `<main>`, `<section>`, `<footer>`).
        * Utilize pelo menos um `<div>` para mostrar que ele não tem significado por si só, mas pode ser usado para agrupar.
    * **Inspiração:** Como a seção "Sobre" ou a "Experiência" do LinkedIn é organizada? Tente replicar algo parecido com HTML.
